```python
from bokeh.plotting import figure, output_file, show, output_notebook
from bokeh.models import ColumnDataSource
```

```python
#Q1. Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
x = np.linspace(0,10,100)
y= np.sin(x)

df = figure(title="Sine wove", x_axis_label='x', y_axis_label='sin(x)')
df.line(x,y, line_width=2)
```

```python
#Q2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
N = 50
x = np.random.random(N)
y = np.random.random(N)
sizes = np.random.randint(10, 100, N)
colors = np.random.choice(['red', 'blue', 'green'], N)
source = ColumnDataSource(data={'x': x, 'y': y, 'sizes': sizes, 'colors': colors})
p = figure(title="Scatter Plot with Sizes and Colors", x_axis_label='x', y_axis_label='y')
p.scatter('x', 'y', size='sizes', color='colors', source=source)
```

```python
#Q3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges, Bananas', 'Pears']
counts = [ 20, 25, 30, 35]

p = figure(x_range = fruits, height=350, title= "Bar char",
           toolbar_location=None, tools="")

p.vbar(x=fruits, top=counts, width=0.9)

p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
```

```python
#Q5.  Create a Bokeh heatmap using the provided dataset.
from bokeh.layouts import layout
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.transform import linear_cmap
output_notebook()
data_heatmap =  np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

x_flat = xx.flatten()
y_flat = yy.flatten()
values = data_heatmap.flatten()

# Create DataFrame from the heatmap data
data = pd.DataFrame({'x': x_flat, 'y': y_flat, 'value': values})

# Define color mapper
colors = ['#440154', '#3b528b', '#21918c', '#5ec962', '#fde725']  # A simple color palette
mapper = LinearColorMapper(palette=colors, low=data['value'].min(), high=data['value'].max())

# Create a figure
p = figure(title="Bokeh Heatmap", x_axis_label='X', y_axis_label='Y',
           tools="hover", tooltips=[('Value', '@value')], width=600, height=600)

# Add heatmap (rectangles for each grid)
p.rect(x='x', y='y', width=1/10, height=1/10, source=data,
       fill_color=linear_cmap('value', palette=colors, low=data['value'].min(), high=data['value'].max()),
       line_color=None)

# Add color bar
color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show plot
show(p)
```
