```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import plotly.graph_objects as go
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')
```

```python
#Q1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a three- dimensional space.
np.random.seed(30)
data = {'X': np.random.uniform(-10, 10, 300),
        'Y': np.random.uniform(-10, 10, 300),
        'Z': np.random.uniform(-10, 10, 300)}
df = pd.DataFrame(data)
```

```python
df1 = go.Figure()
df1.add_trace(go.Scatter3d(x =df.X, y=df.Y, mode = 'markers',z = df['Z'] ))
```

```python
#Q2. Using the Student Grades, create a violin plat to display the distribution of scores across different grade categories.
np.random.seed(15)
data = {'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
        'Score': np.random.randint(50, 100, 200)}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
sns.violinplot(x='Grade', y='Score', data=df)

plt.show()
```

```python
#Q3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
np.random.seed(28)
data = {'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
        'Day': np.random.choice(range (1, 31), 100),
        'Sales': np.random.randint(1000, 5000, 100)}

df = pd.DataFrame(data)

fig = px.imshow(df)
fig.show()
```

```python
#Q4. Using the given x and y data, generate a 3D surface plot to visualize the function
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}

fig = plt.figure(figsize=(14,9))
ax = plt.axes(projection = '3d')

ax.plot_surface(x,y,z)
plt.show()
```

```python
#Q5. Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (x-axis), and bubble size proportional to the population.
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK','Germany', 'France'],
    'Population':np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000,5)
}
df = pd.DataFrame(data)

plt.figure(figsize=(10,6))
plt.scatter(df['GDP'], df['Population'], s=df['Population']*10, alpha=0.7)

for i in range(len(df)):
   plt.text(df['GDP'][i], df['Population'][i], df['Country'][i], fontsize=10)
plt.grid()
plt.show()
```
